```python
def promedio(lista):
    lista = lista.reset_index(drop=True)
    
    if len(lista) == 1:
        return lista[0]
    
    seconds = 0
    for i in range(1,len(lista)):
        seconds += (lista[i]-lista[i-1])

    promedio = seconds/len(lista)
    return promedio

ventana1_promedios = ventana1[['device_id','date_seconds']].groupby('device_id').agg({'date_seconds':promedio})
ventana1_promedios = ventana1_promedios.reset_index().rename(columns={'date_seconds':'prom_segundos'})
ventana1_promedios.head(10)
```
